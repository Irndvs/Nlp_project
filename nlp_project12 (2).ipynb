```python
# Step 1: Install necessary libraries


# Step 2: Import libraries
from transformers import pipeline
import nemo.collections.asr as nemo_asr

# Step 3: Load the ASR model and transcribe the audio file
asr_model = nemo_asr.models.EncDecCTCModel.from_pretrained(model_name="QuartzNet15x5Base-En")
audio_file_path = "/content/d02.wav"
transcript = asr_model.transcribe([audio_file_path])[0]
print("Transcription:", transcript)

# Step 4: Load the emotion recognition model
emotion_classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True)

# Step 5: Recognize emotion in the transcribed text
emotion_predictions = emotion_classifier(transcript)

# Step 6: Print the emotion predictions
print("Emotion Predictions:")
for emotion in emotion_predictions[0]:
    print(f"{emotion['label']}: {emotion['score']:.4f}")
```
